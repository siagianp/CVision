```python
"""
Week 3: Different filters and filtering on varying color spaces

(Some of them are taken from OpenCV tutorials)
"""

from matplotlib import pyplot as plt
import numpy as np
import cv2

# read the image
im_ori = cv2.imread('E:\P-Project & Lab\OpenCV\IR.jpg')
plt.imshow(cv2.cvtColor(im_ori, cv2.COLOR_BGR2RGB))
plt.xticks([]), plt.yticks([])
plt.show()

# smoothing with box filter with normalization
box_filter_norm = np.ones((7, 7), np.float32) / 35
im_box_norm = cv2.filter2D(im_ori, -1, box_filter_norm)

# smoothing with box filter without normalization
box_filter = np.ones((7, 7), np.float32)
im_box = cv2.filter2D(im_ori, -1, box_filter)

# smooth with Gaussian filter, see the difference with box filter
im_gauss = cv2.GaussianBlur(im_ori, (7, 7), 0)

plt.imshow(cv2.cvtColor(im_box, cv2.COLOR_BGR2RGB)), plt.title('Box without norm')
plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(cv2.cvtColor(im_box_norm, cv2.COLOR_BGR2RGB)), plt.title('Box')
plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(cv2.cvtColor(im_gauss, cv2.COLOR_BGR2RGB)), plt.title('Gaussian')
plt.xticks([]), plt.yticks([])
plt.show()

# TODO: create different filter kernels with numpy arrays and see what is the effect!
```

```python
# smoothing on different channels of the BGR image
# on red channel
blurred_r = cv2.GaussianBlur(im_ori[:, :, 2], (15, 15), 0)
im_blurred_r = cv2.merge((im_ori[:, :, 0], im_ori[:, :, 1], blurred_r))

plt.subplot(221), plt.imshow(im_ori[:, :, 2], cmap="gray"), plt.title('Red channel')
plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(blurred_r, cmap="gray"), plt.title('Blurred red channel')
plt.xticks([]), plt.yticks([])
plt.subplot(223), plt.imshow(cv2.cvtColor(im_ori, cv2.COLOR_BGR2RGB)), plt.title('Image')
plt.xticks([]), plt.yticks([])
plt.subplot(224), plt.imshow(cv2.cvtColor(im_blurred_r, cv2.COLOR_BGR2RGB)), plt.title('Blurred Image')
plt.xticks([]), plt.yticks([])
plt.show()

# on green channel
blurred_g = cv2.GaussianBlur(im_ori[:, :, 1], (15, 15), 0)
im_blurred_g = cv2.merge((im_ori[:, :, 0], blurred_g, im_ori[:, :, 2]))

plt.subplot(221), plt.imshow(im_ori[:, :, 1], cmap="gray"), plt.title('Green channel')
plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(blurred_g, cmap="gray"), plt.title('Blurred green channel')
plt.xticks([]), plt.yticks([])
plt.subplot(223), plt.imshow(cv2.cvtColor(im_ori, cv2.COLOR_BGR2RGB)), plt.title('Image')
plt.xticks([]), plt.yticks([])
plt.subplot(224), plt.imshow(cv2.cvtColor(im_blurred_g, cv2.COLOR_BGR2RGB)), plt.title('Blurred Image')
plt.xticks([]), plt.yticks([])
plt.show()

# on blue channel
blurred_b = cv2.GaussianBlur(im_ori[:, :, 0], (15, 15), 0)
im_blurred_b = cv2.merge((blurred_b, im_ori[:, :, 1], im_ori[:, :, 2]))

plt.subplot(221), plt.imshow(im_ori[:, :, 0], cmap="gray"), plt.title('Blue channel')
plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(blurred_b, cmap="gray"), plt.title('Blurred blue channel')
plt.xticks([]), plt.yticks([])
plt.subplot(223), plt.imshow(cv2.cvtColor(im_ori, cv2.COLOR_BGR2RGB)), plt.title('Image')
plt.xticks([]), plt.yticks([])
plt.subplot(224), plt.imshow(cv2.cvtColor(im_blurred_b, cv2.COLOR_BGR2RGB)), plt.title('Blurred Image')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# smoothing on different channels of the Lab image
im_lab = cv2.cvtColor(im_ori, cv2.COLOR_BGR2LAB)

# on luminance channel (intensity)
blurred_l = cv2.GaussianBlur(im_lab[:, :, 0], (15, 15), 0)
im_blurred_l = cv2.cvtColor(cv2.merge((blurred_l, im_lab[:, :, 1], im_lab[:, :, 2])), cv2.COLOR_LAB2BGR)

plt.subplot(221), plt.imshow(im_lab[:, :, 0], cmap="gray"), plt.title('Luminance channel')
plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(blurred_l, cmap="gray"), plt.title('Blurred luminance channel')
plt.xticks([]), plt.yticks([])
plt.subplot(223), plt.imshow(cv2.cvtColor(im_ori, cv2.COLOR_BGR2RGB)), plt.title('Image')
plt.xticks([]), plt.yticks([])
plt.subplot(224), plt.imshow(cv2.cvtColor(im_blurred_l, cv2.COLOR_BGR2RGB)), plt.title('Blurred Image')
plt.xticks([]), plt.yticks([])
plt.show()

# on "a" channel (green to red)
blurred_a = cv2.GaussianBlur(im_lab[:, :, 1], (15, 15), 0)
im_blurred_a = cv2.cvtColor(cv2.merge((im_lab[:, :, 0], blurred_a, im_lab[:, :, 2])), cv2.COLOR_LAB2BGR)

plt.subplot(221), plt.imshow(im_lab[:, :, 1], cmap="gray"), plt.title('"a" channel')
plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(blurred_a, cmap="gray"), plt.title('Blurred "a" channel')
plt.xticks([]), plt.yticks([])
plt.subplot(223), plt.imshow(cv2.cvtColor(im_ori, cv2.COLOR_BGR2RGB)), plt.title('Image')
plt.xticks([]), plt.yticks([])
plt.subplot(224), plt.imshow(cv2.cvtColor(im_blurred_a, cv2.COLOR_BGR2RGB)), plt.title('Blurred Image')
plt.xticks([]), plt.yticks([])
plt.show()

# on "b" channel (blue to yellow)
blurred_b = cv2.GaussianBlur(im_lab[:, :, 2], (15, 15), 0)
im_blurred_b = cv2.cvtColor(cv2.merge((im_lab[:, :, 0], im_lab[:, :, 1], blurred_b)), cv2.COLOR_LAB2BGR)

plt.subplot(221), plt.imshow(im_lab[:, :, 2], cmap="gray"), plt.title('"b" channel')
plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(blurred_b, cmap="gray"), plt.title('Blurred "b" channel')
plt.xticks([]), plt.yticks([])
plt.subplot(223), plt.imshow(cv2.cvtColor(im_ori, cv2.COLOR_BGR2RGB)), plt.title('Image')
plt.xticks([]), plt.yticks([])
plt.subplot(224), plt.imshow(cv2.cvtColor(im_blurred_b, cv2.COLOR_BGR2RGB)), plt.title('Blurred Image')
plt.xticks([]), plt.yticks([])
plt.show()

# TODO: apply your filters on different channels of Lab color space to see their effect!
```
